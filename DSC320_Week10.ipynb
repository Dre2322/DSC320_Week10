```python
# 1. Summarizing Data
# a

import pandas as pd

# read a csv file
df = pd.read_csv('qb_stats.csv')

# drop the first and second column
df = df.drop(df.columns[[0, 1]], axis=1)

# calculate the mean of the remaining columns
mean = df.mean()

# print the mean
print(mean)
```

```python
# b

# calculate the standard deviation of the remaining columns
std_dev = df.std()

# print the standard deviation
print(std_dev)
```

```python
# c

import matplotlib.pyplot as plt

# create a histogram of the yds column
plt.hist(df['yds'], bins=50, color='skyblue')

# label the axes
plt.xlabel('yards')
plt.ylabel('count')

# show the plot
plt.show()
```

```python
# d

import numpy as np

# pull tds column from df
td = df['td']

# create a boxplot
plt.boxplot(td)

# find the first quartile
q1 = np.percentile(td, 25)

# find the third quartile
q3 = np.percentile(td, 75)

# find the interquartile range
iqr = q3 - q1

# find the lower and upper whiskers
lower_whisker = q1 - (1.5 * iqr)
upper_whisker = q3 + (1.5 * iqr)

# find the outliers
outliers = td[(td <= lower_whisker) | (td >= upper_whisker)] 

# print the outliers
print(f'Outliers: {outliers}')

# add a title
plt.title('Boxplot of Touchdowns')

# add x and y labels
plt.xlabel('Touchdowns')
plt.ylabel('Number of Touchdowns')

# show the plot
plt.show()
```

```python
# 2. Calculating Probabilities from Data
# a

import pandas as pd

# load survey_data.csv into numpy array
data = pd.read_csv('survey_data.csv')

# calculate the total number of respondents
total = len(data)

# count number of brown hair in the data
brown_hair = len(data[data['hair_color'] == 'brown'])

# calculate the probability of brown hair 
brown_hair_prob = brown_hair / total

# print the probability of brown hair
print(f'The probability of a college student has brown hair is: \n{brown_hair_prob}')

```

```python
# b

# count number of blue eyes in the data
blue_eyes = len(data[data['eye_color'] == 'blue'])

# calculate the probability of blue eyes 
blue_eyes_prob = blue_eyes / total

# print the probability of blue eyes
print(f'The probability of a college student has blue eye is: \n{blue_eyes_prob}')
```

```python
# c

# count the number of students with brown hair and blue eyes
brown_hair = len(data[data['hair_color'] == 'brown'])
blue_eyes = len(data[data['eye_color'] == 'blue'])

# calculate the percentage of students with brown hair and blue eyes
brown_hair_percentage = (brown_hair / total) * 100
blue_eyes_percentage = (blue_eyes / total) * 100

# find the probability a college student has blue eyes given that they have brown hair
probability = (blue_eyes_percentage / brown_hair_percentage) * 100

# print the results
print('The probability a college student has blue eyes given that they have brown hair is:')
print(f'{probability:.2f}%')
```

```python
# d

# find the probability a college student has brown hair given that they have blue eyes
probability = (brown_hair_percentage * blue_eyes_percentage) / 100

# print the results
print('The probability a college student has brown hair given that they have blue eyes is:')
print(f'{probability:.2f}%')
```

```python
# e
# yes the results indicate students having brown hair and blue eyes are independent from one another.
# when comparing the students with blue eyes given they have brown hair, you are dealing with a smaller 
# percentage than when students have brown hair, given they have blue eyes.
```

```python
# f

import matplotlib.pyplot as plt

# load survey_data.csv into numpy array
data = pd.read_csv('survey_data.csv')

# seperate out the data into two arrays
hair_color = data['hair_color']
eye_color = data['eye_color']

# count the number of each unique hair color
hair_color_count = pd.value_counts(hair_color)

# count the number of each unique eye color
eye_color_count = pd.value_counts(eye_color)

# create a double bar graph of hair color and eye color
plt.bar(hair_color_count.index, hair_color_count.values, color='blue')
plt.bar(eye_color_count.index, eye_color_count.values, color='green')

# add labels and title
plt.xlabel('Hair and Eye Color')
plt.ylabel('Count')
plt.title('Hair and Eye Color Survey Results')

# add legened
plt.legend(['Hair Color', 'Eye Color'])

# show the graph
plt.show()
```
